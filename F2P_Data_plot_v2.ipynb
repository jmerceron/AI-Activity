```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from the Excel file
years = [2029, 2030, 2031, 2032, 2033, 2034, 2035, 2036, 2037]

data = {
    'High': {
        'installs': [20, 125, 250, 350, 400, 500, 600, 650, 650],
        'mau': [20, 50, 60, 70, 80, 80, 70, 20, 15],
        'monthly_paying': [0.5, 1.2, 1.5, 1.8, 2.0, 1.8, 1.5, 1.0, 0.8],
        'arppu': [80, 100, 110, 150, 160, 155, 150, 140, 130]
    },
    'Mid': {
        'installs': [10, 70, 125, 175, 200, 250, 250, 325, 325],
        'mau': [10, 25, 30, 35, 40, 40, 35, 10, 7],
        'monthly_paying': [0.2, 0.6, 0.7, 0.9, 1.0, 0.8, 0.7, 0.5, 0.4],
        'arppu': [40, 50, 55, 70, 80, 75, 70, 70, 60]
    },
    'Low': {
        'installs': [5, 35, 60, 65, 100, 125, 125, 150, 150],
        'mau': [5, 12, 15, 17, 20, 20, 17, 5, 3],
        'monthly_paying': [0.1, 0.3, 0.35, 0.45, 0.5, 0.4, 0.35, 0.25, 0.2],
        'arppu': [20, 25, 27, 35, 40, 37, 35, 35, 30]
    }
}

# Calculate gross revenue (in millions)
for projection in data:
    data[projection]['gross_revenue'] = [round(mpu * 12 * arppu / 1000, 1) for mpu, arppu in
                                        zip(data[projection]['monthly_paying'], data[projection]['arppu'])]

# Create the plots
plt.figure(figsize=(18, 12))

# 1. Gross Revenue Evolution
plt.subplot(3, 1, 1)
for projection in data:
    plt.plot(years, data[projection]['gross_revenue'], marker='o', label=f'LEGION {projection}')
plt.title('Gross Revenue Evolution (in $M)')
plt.ylabel('Revenue ($M)')
plt.grid(True)
plt.legend()

# 2. Installs Evolution
plt.subplot(3, 1, 2)
for projection in data:
    plt.plot(years, data[projection]['installs'], marker='o', label=f'LEGION {projection}')
plt.title('Installs Evolution (in millions)')
plt.ylabel('Installs (M)')
plt.grid(True)
plt.legend()

# 3. MAU Evolution
plt.subplot(3, 1, 3)
for projection in data:
    plt.plot(years, data[projection]['mau'], marker='o', label=f'LEGION {projection}')
plt.title('Monthly Active Users Evolution (in millions)')
plt.ylabel('MAU (M)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
