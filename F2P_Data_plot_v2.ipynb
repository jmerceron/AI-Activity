```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from the Excel file
years = [2029, 2030, 2031, 2032, 2033, 2034, 2035, 2036, 2037]

projections = {
    'HIGH': {
        'installs': [10, 70, 125, 175, 200, 250, 250, 325, 325],
        'yau': [7, 22, 27, 32, 35, 40, 30, 15, 12],
        'arpu': [10, 17, 22, 30, 40, 35, 32, 30, 28],
        'mau': [10, 25, 30, 35, 40, 40, 35, 10, 7]
    },
    'MID': {
        'installs': [5, 35, 60, 65, 100, 125, 125, 150, 150],
        'yau': [3, 11, 14, 15, 16, 20, 15, 7, 6],
        'arpu': [5, 8, 11, 15, 20, 17, 16, 15, 14],
        'mau': [5, 12, 15, 17, 20, 20, 17, 5, 3]
    },
    'LOW': {
        'installs': [2.5, 15, 30, 32, 50, 60, 60, 70, 70],
        'yau': [1.5, 5, 7, 7, 8, 10, 0.7, 3.5, 3],
        'arpu': [2.5, 4, 5, 7, 10, 8, 8, 7, 6],
        'mau': [2.5, 5, 7, 8, 10, 10, 8, 2.5, 1.5]
    },
    'VERY LOW': {
        'installs': [1.2, 7, 15, 15, 25, 30, 30, 35, 35],
        'yau': [0.7, 2.5, 3.5, 3.5, 4, 5, 0.35, 1.5, 1.5],
        'arpu': [1.5, 2, 3, 3, 5, 4, 4, 3, 2],
        'mau': [1, 2, 3, 4, 5, 5, 4, 1, 1]
    }
}

# Calculate gross revenue (in millions)
for proj in projections:
    projections[proj]['gross_revenue'] = [yau * arpu for yau, arpu in zip(projections[proj]['yau'], projections[proj]['arpu'])]

# Create the plots
plt.figure(figsize=(18, 12))

# 1. Gross Revenue Evolution
plt.subplot(3, 1, 1)
for proj in projections:
    plt.plot(years, projections[proj]['gross_revenue'], label=f'{proj}', marker='o')
plt.title('LEGION Mobile - Gross Revenue Evolution (2029-2037)')
plt.ylabel('Gross Revenue ($M)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.xticks(years)

# 2. Installs Evolution
plt.subplot(3, 1, 2)
for proj in projections:
    plt.plot(years, projections[proj]['installs'], label=f'{proj}', marker='o')
plt.title('LEGION Mobile - Installs Evolution (2029-2037)')
plt.ylabel('Installs (M)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.xticks(years)

# 3. MAU Evolution
plt.subplot(3, 1, 3)
for proj in projections:
    plt.plot(years, projections[proj]['mau'], label=f'{proj}', marker='o')
plt.title('LEGION Mobile - MAU Evolution (2029-2037)')
plt.ylabel('Monthly Active Users (M)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.xticks(years)

plt.tight_layout()
plt.show()
```
